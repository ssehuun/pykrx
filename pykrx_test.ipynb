## 일자별 OHLCV 조회

#### get_market_ohlcv_by_date() 함수는 시작일/종료일/티커 세 개의 파라미터를 입력받아 OHLCV를 일자별로 정렬하여 DataFrame으로 반환합니다.

```python
df = stock.get_market_ohlcv_by_date("20201201", "20201221", "000660")
print(df.head(3))
```

#### get_market_ohlcv_by_date() 함수에 frequency 파라미터(d/m/y)를 추가할 수 있습니다. d는 일자, m은 월, y는 연도를 의미합니다. 다음은 20180810부터 20181212까지의 삼성전자 월별 OHLCV 데이터를 출력합니다.

```python
df = stock.get_market_ohlcv_by_date("20201001", "20201221", "005930", "m")
print(df.head(3))
```

#### get_stock_ticker_list() 함수와 함께 사용하면 전종목의 OHLCV를 가져올 수 있습니다. 한 번에 너무 많은 정보를 요청하면 KRX 서버가 차단할 수 있으므로 time 모듈을 사용해서 1초 지연 후 요청하도록 코드를 작성했습니다.

```python
import time
#for ticker in stock.get_stock_ticker_list(): # AttributeError: module 'pykrx.stock' has no attribute 'get_stock_ticker_list'
for ticker in stock.get_market_ticker_list():
    df = stock.get_market_ohlcv_by_date("20201218", "20201221", ticker)
    print(df.head())
    time.sleep(1)
```

## 2.1.3 공매도 API

```python
# 인자 4개 함수는 정의되어 있지 않음............
df = stock.get_etf_ohlcv_by_date("20190401", "20190405", "295820", 'm')
print(df.head())
```

```python
import time
tickers = stock.get_etf_ticker_list()
for ticker in tickers[:10]:    
    df = stock.get_etf_ohlcv_by_date("20190301", "20190405", ticker)
    print(df.head(2))
    time.sleep(1)
```
